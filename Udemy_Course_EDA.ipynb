# EDA on Levels

## Levels

- How many levels do we have

- What is the distribution of courses per levels

- Which subject have the highest levels

- How many subscribers per levels

- How many courses per levels


```python
# How many levels do we have
ds['level'].unique()
```

```python
# What is the distribution of courses per levels
ds['level'].value_counts()
```

```python
# What is the distribution of courses per levels
ds['level'].value_counts().plot(kind='bar' , color = 'green')
plt.title('Distribution of levels')
plt.show()
```

```python
#  How many subscribers per levels
ds.groupby('level')['num_subscribers'].sum()
```

```python
ds[['num_subscribers','level']]
```

```python
# Plot of How many subscribers per levels
ds.groupby('level')['num_subscribers'].sum().plot(kind='bar' , color = 'Green')
plt.title("Subcribers per level")
plt.show()
```

```python
# How many levels per subject category
ds.groupby('subject')['level'].value_counts()
```

```python
# Plot of How many levels per subject category
plt.figure(figsize=(20,10))
sns.barplot(x='level',y='num_subscribers', hue='subject',data=ds  )
plt.show()
```

# Correlation Question
+ Correlation Questions
	- Does number of subscribers depend on
	- number of reviews
	- price
	- number of lectures
	- content duration

```python
sns.scatterplot(data=ds,x='price',y='num_subscribers')
```

```python
sns.scatterplot(data=ds,x='price',y='num_reviews')
```

```python
plt.figure(figsize=(20,10))
plt.title("Does Price Influence Subscription Per Subject Category")
sns.lineplot(data=ds,x='price',y='num_subscribers',hue='subject')
plt.show()
```
